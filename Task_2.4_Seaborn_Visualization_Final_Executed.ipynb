```python
import seaborn as sns
import matplotlib.pyplot as plt

# Apply a modern base theme
sns.set_theme(style='whitegrid')

# Use a cool, consistent palette for all charts
sns.set_palette('crest')  # or try 'rocket_r', 'mako', or 'viridis' for variety
```

```python
# Set theme
sns.set_theme(style="whitegrid", palette="viridis")
```

```python
# Sample data creation (simulate a small working set for the notebook)
data = {
    'date': pd.date_range(start='2022-01-01', periods=10, freq='D'),
    'start_station_name': ['Station A', 'Station B', 'Station C', 'Station D', 'Station E',
                           'Station A', 'Station B', 'Station C', 'Station D', 'Station E'],
    'trip_count': [100, 150, 120, 130, 80, 170, 180, 110, 160, 140],
    'temperature': [2, 3, 4, 5, 3, 6, 7, 4, 8, 5],
    'user_type': ['Subscriber', 'Customer', 'Subscriber', 'Customer', 'Subscriber',
                  'Customer', 'Subscriber', 'Customer', 'Subscriber', 'Customer']
}
df = pd.DataFrame(data)
```

```python
# --- Bar Chart of Top Start Stations ---
plt.figure(figsize=(10, 6))
top_stations = df['start_station_name'].value_counts()
sns.barplot(x=top_stations.values, y=top_stations.index, palette='crest')
plt.title("Top Starting Stations by Frequency")
plt.xlabel("Trip Count")
plt.ylabel("Start Station")
plt.tight_layout()
plt.show()
```

```python
# --- Dual-Axis Line Chart ---
fig, ax1 = plt.subplots(figsize=(12, 6))
sns.lineplot(x='date', y='trip_count', data=df, ax=ax1, color='red', label="Trip Count")
ax1.set_ylabel("Trip Count", color='red')
ax1.tick_params(axis='y', labelcolor='red')
```

```python
ax2 = ax1.twinx()
sns.lineplot(x='date', y='temperature', data=df, ax=ax2, color='red', label="Temperature")
ax2.set_ylabel("Temperature (C)", color='red')
ax2.tick_params(axis='y', labelcolor='red')
```

```python
ax1.set_title("Trip Count vs Temperature Over Time")
fig.tight_layout()
plt.show()
```

```python
# --- Box Plot by User Type (Improved) ---
plt.figure(figsize=(8, 6))
sns.boxplot(
    x='user_type',
    y='trip_count',
    data=df,
    hue=None,
    palette=['#4C72B0', '#55A868']  # soft blue and green shades for contrast
)
plt.title("Trip Count Distribution by User Type", fontsize=14, fontweight='bold', color='navy')
plt.xlabel("User Type", fontsize=12)
plt.ylabel("Trip Count", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

**Box Plot Analysis:**
- The median trip count varies between user types.
- Customers show slightly higher variability in trip counts.
- Subscribers have more consistent usage patterns, while customers show wider spread in values.

```python
# --- FacetGrid Boxplot ---
g = sns.FacetGrid(df, col="user_type", height=5, aspect=1.2)
g.map_dataframe(sns.boxplot, x="start_station_name", y="trip_count", palette="coolwarm")
g.set_xticklabels(rotation=45)
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle("Trip Count by Station and User Type")
plt.show()
```

**FacetGrid Insights:**
- The distribution of trip counts by station is clearer when split by user type.
- This plot helps reveal station popularity differences for each user group.
